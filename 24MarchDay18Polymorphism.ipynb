```python
# Polymorphism
# Method Overriding(through the Inheritance)
# one method with same name
class Animal:  # parent class 
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")
a = Animal()
a.sound()
```

```python
# method overloading: same name with different parameter list
class math():
    def add(self,x,y): # default arguments
        return x+y
    def add(self,x,y,z):
        return x+y+z
    def add(self,a,b,c,d):
        return a+b+c+d
m = math()
print(m.add(10,5,0,0))
print(m.add(5,6,7,0))
print(m.add(5,6,7,8))  # last defined method only stored remaining not stored
```

```python
# Operator Overloading
# Here, + is overloaded to add marks.
class Student:
    def __init__(self, marks):
        self.marks = marks
    
    def __add__(self, other):
        return self.marks + other.marks

s1 = Student(80)
s2 = Student(70)

print(s1 + s2)  
```

```python
# More Operators
class Demo:
    def __init__(self, x):
        self.x = x
    
    def __sub__(self, other):
        return self.x - other.x

a = Demo(10)
b = Demo(5)

print(a - b)   
```

```python
# Duck Typing
class Dog:
    def speak(self):
        return "Bark"

class Cat:
    def speak(self):
        return "Meow"

def make_sound(animal):
    print(animal.speak())

d = Dog()
c = Cat()

make_sound(d) 
make_sound(c)  
```

```python
# Another Example
# Both work because they have move() method
class Car:
    def move(self):
        print("Car is moving")

class Bike:
    def move(self):
        print("Bike is moving")

def start(vehicle):
    vehicle.move()

start(Car())
start(Bike())
```
