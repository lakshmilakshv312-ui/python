```python
# Types Of Attributes
# 1. Class Attributes
# Class Defination
class Student:
    school_name = "ZP high school" # Class Attribute
    def __init__(self, name, age): # -------------------> Initilization Constructor
        self.name = name # Instance Attribute
        self.age = age  # Instance Attribute
    def details(self):  # Method
        return f"My name is {self.name} and I am a {self.age} years old"
# Object Creation
s1 = Student("venu", 23) # s1 and s2 are objects
s2 = Student("lakshmi", 21)
print(s1.name)
print(s2.name)
print(s1.details())
print(s2.details())
print(s1.school_name)
print(s2.school_name)
```

```python
# caluclator program include add, sub, mul, div, and create 2 objects
class Calculator:
    def add(self, a, b):
        return a + b
    def sub(self, a, b):
        return a - b
    def mul(self, a, b):
        return a * b
    def div(self, a, b):
        if b != 0:
            return a / b
        else:
            return "Cannot divide by zero"
# Creating 2 objects
calc1 = Calculator()
calc2 = Calculator()

# Taking input
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# Using first object
print("\nUsing calc1:")
print("Addition:", calc1.add(a, b))
print("Subtraction:", calc1.sub(a, b))

# Using second object
print("\nUsing calc2:")
print("Multiplication:", calc2.mul(a, b))
print("Division:", calc2.div(a, b))
```

```python
# Method calling using class & object
class Student:
    school_name = "ZP high school"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def details(self):
        return f"{self.name} - {self.age}"

s1 = Student("venu", 23)
s2 = Student("lakshmi", 21)

# Calling using object
print(s1.details())

# Calling using class
print(Student.details(s2))
```

```python
# Method with arguments (parameters)
class Student:
    school_name = "ZP high school"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def update_age(self, new_age):
        self.age = new_age

    def details(self):
        return f"{self.name} is {self.age} years old"

s1 = Student("venu", 23)
s2 = Student("lakshmi", 21)

s1.update_age(25)
s2.update_age(22)

print(s1.details())
print(s2.details())
```
