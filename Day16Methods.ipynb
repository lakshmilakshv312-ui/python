```python
# Types of methods
# 1. Instance method
# 2. class method
# 3. static method
class Student:
    school_name = "ZP high school" # Class Attribute
    def __init__(self, name, age): # -------------------> Initilization Constructor
        self.name = name # Instance Attribute
        self.age = age  # Instance Attribute
    def details(self):  # instance Method
        return f"My name is {self.name} and I am a {self.age} years old"

    @classmethod  
    def change_school_name(cls, new_name): # class method # we can call class name along with object name
        cls.school_name = new_name

    @staticmethod 
    def percentage(m1,m2,m3): # static method
        return(m1 + m2 + m3)/3
# Object Creation
s1 = Student("venu", 23) # s1 and s2 are objects
s2 = Student("lakshmi", 21)
print(s1.age)
s1.change_school_name("XYZ School")
print(s1.school_name)
print(s2.school_name)
s1.change_school_name("ABC School")
print(s1.school_name)
print(s2.school_name)
print(s1.percentage(85,86,87))
print(Student.percentage(85,86,87))
```

```python
# Encapsulation
# 1. public
# 2. protected
# 3. private
# create bank class
class Bank:
    def __init__(self, name, email, amount=1000):
        self.name = name  # private variable
        self._email = email
        self.__amount = amount
        print("Constructor exucated")

    # string method
    def __str__(self):
       return f"The user name is {self.name} and email address is {self._email}"
    def withdraw(self,withdraw_amount):
        if self.__amount >= withdraw_amount:
            self.__amount-= withdraw_amount
            return f"withdrawal successful and cuurent balance is {self.__amount}"
        return "Insufficient balance"
    def __Total_User(self):
        return "Total users in bank is 102"

account1 = Bank("venu", "venuroyal@wipro.com", 10000)
print(account1)
print(account1.name)
print(account1.withdraw(1000))
print(account1._email)# protected member access in out side the class
#print(account1._Bank__Total_Users()) # private member access by using "name mangling"
```

```python
class Bank:
    def __init__(self, name, balance, pin):
        self.name = name          # Public variable
        self._balance = balance   # Protected variable
        self.__pin = pin          # Private variable

    # Public method
    def show_details(self):
        print("Name:", self.name)
        print("Balance:", self._balance)

    # Protected method
    def _deposit(self, amount):
        self._balance += amount
        print("Deposited:", amount)

    # Private method
    def __withdraw(self, amount, pin):
        if pin == self.__pin:
            self._balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Wrong PIN")

    # Access private method using public method
    def withdraw_money(self, amount, pin):
        self.__withdraw(amount, pin)

# Object creation
b1 = Bank("Likitha", 10000, 1234)

# Access public
b1.show_details()

# Access protected (possible but not recommended)
b1._deposit(2000)

# Access private (not directly possible)
# b1.__withdraw(1000, 1234) ❌ Error

# Correct way to access private
b1.withdraw_money(1000, 1234)

# Final balance
b1.show_details()
```
