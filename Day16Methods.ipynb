```python
# Types of methods
# 1. Instance method
# 2. class method
# 3. static method
class Student:
    school_name = "ZP high school" # Class Attribute
    def __init__(self, name, age): # -------------------> Initilization Constructor
        self.name = name # Instance Attribute
        self.age = age  # Instance Attribute
    def details(self):  # instance Method
        return f"My name is {self.name} and I am a {self.age} years old"

    @classmethod  
    def change_school_name(cls, new_name): # class method # we can call class name along with object name
        cls.school_name = new_name

    @staticmethod 
    def percentage(m1,m2,m3): # static method
        return(m1 + m2 + m3)/3
# Object Creation
s1 = Student("venu", 23) # s1 and s2 are objects
s2 = Student("lakshmi", 21)
print(s1.age)
s1.change_school_name("XYZ School")
print(s1.school_name)
print(s2.school_name)
s1.change_school_name("ABC School")
print(s1.school_name)
print(s2.school_name)
print(s1.percentage(85,86,87))
print(Student.percentage(85,86,87))
```
