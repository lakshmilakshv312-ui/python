```python
password = input()
lower_ind = False
upper_ind = False
digit_ind = False
special_ind = False
length = len(password)>= 8
for char in password:
    if char.isupper():
        upper_ind = True
    elif char.islower():
        lower_ind = True
    elif char.isdigit():
        digit_ind = True
    elif char in "@#$.!^":
        special_ind = True
if lower_ind and upper_ind and digit_ind and special_ind:
    print("Strong password")
else:
    print("Weak password")
```
