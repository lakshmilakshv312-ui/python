```python
# set operations or methods
# 1. union(new_set): it returns the all unique elements from sets
s1 = {1,2,3,6}
s2 = {3,4,5,6}
print(s1.union(s2))
```

```python
# 2. intersection(new_set): it returns only common elemenmts
s1 = {1,2,3,6}
s2 = {3,4,5,6}
print(s1 | s2)
```

```python
# 3. diffrences: Returns elements in one set but not in the other
s1 = {1,2,3,6}
s2 = {3,4,5,6}
print(s1.difference(s2))
print(s1-s2)
```

```python
# 4. symmetric difference: Returns elements in either set but not in both
s1 = {1,2,3,6}
s2 = {3,4,5,6}
print(s1.symmetric_difference(s2))
```

```python
# 5. subset: A subset means all elements of one set are present in another set.
s1 = {1,2,3,6}
s2 = {3,4,5,6}
print("subset")
print(s1.issubset(s2))
print({1,2}.issubset(s1))
print({1,2,5}.issubset(s1))
print(set().issubset(s1))# every empty set is a subset
```

```python
# 6. super set: A superset means a set contains all elements of another set.
s1 = {1,2,3,6}
s2 = {3,4,5,6}
print("super set")
print(s1.issuperset(s2))
print({1,2}.issuperset(s1))
print({1,2,5}.issuperset(s1))
print(set().issuperset(s1))
```

```python
# 7. disjoint set: A disjoint set means two sets that have no common elements.
s1 = {1,2,3,6}
s2 = {3,4,5,6}
print("super set")
print(s1.isdisjoint(s2))
print({1,2}.isdisjoint(s1))
print({1,2,5}.isdisjoint(s1))
print(set().isdisjoint(s1))
```

```python
# add(ele): it add to set, if the element is not present in set, otherwise nothing
# update(new_set): set elments updates with new set elements
# remove(ele); it removes the set element, if element not present in set it raises keyError
# discard(ele): it removes the element from set otherwise nothing.
s = {1,2,3,4}
s.add(10)
print(s)
s.update({1,2,5,6})
print(s)
s.remove(10)
print(s)
# s.remove(20) # keyError
# print(s)
s.discard(100)
print(s)
s.discard(5) # it removes the element from set otherwise nothing 
print(s)
```
