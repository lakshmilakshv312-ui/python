```python
# Multilevel Inheritance
# Multilevel inheritance is an OOP concept where a class inherits from another class, and then another class inherits from that derived class.
# Animal Example
class Animal:
    def eat(self):
        print("Animal is eating")

class Dog(Animal):
    def bark(self):
        print("Dog is barking")

class Puppy(Dog):
    def weep(self):
        print("Puppy is weeping")

p = Puppy()
p.eat()
p.bark()
p.weep()
```

```python
# Multilevel Inheritance
# Vehicle Example
class Vehicle:
    def start(self):
        print("Vehicle starts")

class Car(Vehicle):
    def drive(self):
        print("Car is driving")

class ElectricCar(Car):
    def charge(self):
        print("Electric car is charging")

e = ElectricCar()
e.start()
e.drive()
e.charge()
```

```python
# Multilevel Inheritance
# Person Example
class Person:
    def show(self):
        print("I am a person")

class Student(Person):
    def study(self):
        print("I am studying")

class GraduateStudent(Student):
    def research(self):
        print("I am doing research")

g = GraduateStudent()
g.show()
g.study()
g.research()
```

```python
# Multilevel Inheritance
# Company Example
class Company:
    def work(self):
        print("Company work")

class Manager(Company):
    def manage(self):
        print("Managing team")

class TeamLead(Manager):
    def lead(self):
        print("Leading project")

t = TeamLead()
t.work()
t.manage()
t.lead()
```

```python
# Multiple Inheritance
# Multiple inheritance means one class inherits from more than one parent class.
# Basic Example
class Father:
    def skills(self):
        print("Father: Gardening")

class Mother:
    def skills(self):
        print("Mother: Cooking")

class Child(Father, Mother):
    def skills_child(self):
        print("Child: Learning skills")

c = Child()
c.skills()        # Which one will be called?
c.skills_child()
```

```python
# Multiple Inheritance
# Different Methods
class A:
    def showA(self):
        print("Class A")

class B:
    def showB(self):
        print("Class B")

class C(A, B):
    def showC(self):
        print("Class C")

obj = C()
obj.showA()
obj.showB()
obj.showC()
```

```python
# Multiple Inheritance
# Real-Time Example
class Employee:
    def work(self):
        print("Working")

class Student:
    def study(self):
        print("Studying")

class Intern(Employee, Student):
    def role(self):
        print("Intern role")

i = Intern()
i.work()
i.study()
i.role()
```

```python
# Hierarchical Inheritance
# Hierarchical inheritance means one parent class is inherited by multiple child classes.
# Basic Example
class Animal:
    def eat(self):
        print("Animal is eating")

class Dog(Animal):
    def bark(self):
        print("Dog is barking")

class Cat(Animal):
    def meow(self):
        print("Cat is meowing")

d = Dog()
c = Cat()

d.eat()
d.bark()

c.eat()
c.meow()
```

```python
# Hierarchical Inheritance
# Student Example
class Person:
    def show(self):
        print("I am a person")

class Student(Person):
    def study(self):
        print("Studying")

class Teacher(Person):
    def teach(self):
        print("Teaching")

s = Student()
t = Teacher()

s.show()
s.study()

t.show()
t.teach()
```

```python
# Hierarchical Inheritance
# Vehicle Example
class Vehicle:
    def start(self):
        print("Vehicle starts")

class Car(Vehicle):
    def drive(self):
        print("Car driving")

class Bike(Vehicle):
    def ride(self):
        print("Bike riding")

car = Car()
bike = Bike()

car.start()
car.drive()

bike.start()
bike.ride()
```

```python
# Hybrid Inheritance
# Hybrid inheritance is a combination of two or more types of inheritance
# (like multiple + multilevel + hierarchical).
# 1. Multilevel + Multiple
class A:
    def showA(self):
        print("A")

class B(A):
    def showB(self):
        print("B")

class C(A):
    def showC(self):
        print("C")

class D(B, C):
    def showD(self):
        print("D")

obj = D()
obj.showA()
obj.showB()
obj.showC()
obj.showD()
```

```python
# Hybrid Inheritance
# 2. Hierarchical + Multiple
class A:
    def showA(self):
        print("A")

class B(A):
    def showB(self):
        print("B")

class C(A):
    def showC(self):
        print("C")

class D(B, C):
    def showD(self):
        print("D")

d = D()
d.showA()
d.showB()
d.showC()
d.showD()
```

```python
# Hybrid Inheritance
# 3. Single + Multiple
class A:
    def showA(self):
        print("A")

class C:
    def showC(self):
        print("C")

class B(A, C):
    def showB(self):
        print("B")

obj = B()
obj.showA()
obj.showC()
obj.showB()
```

```python
# Hybrid Inheritance
# 4. Multilevel + Hierarchical
class A:
    def showA(self):
        print("A")

class B(A):
    def showB(self):
        print("B")

class C(B):
    def showC(self):
        print("C")

class D(B):
    def showD(self):
        print("D")

c = C()
d = D()

c.showA()
c.showB()
c.showC()

d.showA()
d.showB()
d.showD()
```
