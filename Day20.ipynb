```python
# 11. Employee Payroll System (Inheritance + Polymorphism)
from abc import ABC, abstractmethod

# Base Class
class Employee(ABC):
    def __init__(self, emp_id, name):
        self.emp_id = emp_id
        self.name = name

    @abstractmethod   #optional 
    def calculate_salary(self):
        pass

# Salaried Employee
class SalariedEmployee(Employee):
    def __init__(self, emp_id, name, monthly_salary):
        super().__init__(emp_id, name)
        self.monthly_salary = monthly_salary

    def calculate_salary(self):
        return self.monthly_salary

# Hourly Employee
class HourlyEmployee(Employee):
    def __init__(self, emp_id, name, hourly_rate, hours_worked):
        super().__init__(emp_id, name)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        return self.hourly_rate * self.hours_worked

# Commission Employee
class CommissionEmployee(Employee):
    def __init__(self, emp_id, name, base_salary, commission_rate, total_sales):
        super().__init__(emp_id, name)
        self.base_salary = base_salary
        self.commission_rate = commission_rate
        self.total_sales = total_sales

    def calculate_salary(self):
        return self.base_salary + (self.commission_rate * self.total_sales)

# Testing
emp1 = SalariedEmployee(1, "Venu", 30000)
emp2 = HourlyEmployee(2, "Lakshmi", 200, 120)
emp3 = CommissionEmployee(3, "Likitha", 15000, 0.1, 50000)

employees = [emp1, emp2, emp3]

for emp in employees:
    print(f"{emp.name} Salary: {emp.calculate_salary()}")
```

```python
# 12. Multi-Account Banking System (Inheritance + Polymorphism + Exception Handling)
from abc import ABC, abstractmethod

# Abstract Class
class Account(ABC):
    def __init__(self, acc_no, holder_name, balance):
        self.acc_no = acc_no
        self.holder_name = holder_name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: {amount}")

    @abstractmethod
    def withdraw(self, amount):
        pass

    def get_balance(self):
        return self.balance

# Savings Account
class SavingsAccount(Account):
    MIN_BALANCE = 1000
    INTEREST-_RATE = 0.06

    def withdraw(self, amount):
        if self.balance - amount < self.MIN_BALANCE:
            raise Exception("Minimum balance must be maintained!")
        self.balance -= amount
        print(f"Withdrawn: {amount}")

    def apply_interest(self):
        interest = self.balance * self.INTEREST_RATE
        self.balance += interest
        print(f"Interest added: {interest}")

# Testing
try:
    acc = SavingsAccount(101, "Lakshmi", 5000)

    acc.deposit(2000)
    acc.withdraw(3000)
    acc.apply_interest()

    print("Final Balance:", acc.get_balance())

except Exception as e:
    print("Error:", e)
```
