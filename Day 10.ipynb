```python
# list methods
# adding elements to list
# oppend(ele): Adds the element at end of the list
# extend(list): Adds all elements in list at end
l = [1,2,3,4,5]
l.append(10)
print(l)
l.extend([10,20,30])
print(l)
```

```python
# list methods
# adding elements to list
# oppend(ele): Adds the element at end of the list
# extend(list): Adds all elements in list at end
# insert(pos,ele): insert the element at specified position
l = [1,2,3,4,5]
l.append(10)
print(l)
l.extend([10,20,30])
print(l)
l.insert(3,100)
print(l)
l.insert(100,1000)
print(l)
l.insert(-2,200)
print(l)
```

```python
# Removing elements from list
# remove(val): it removes the first accurence value from list, otherwise ValueError
# pop(): it remove and return end value from list, if list is empty at raises error
# pop(position): it remove and return specified element from list
# clear(): it removes all elements from the list.
l = [1,2,4,5,2]
l.remove(4)
#l.remove(20) # value error
print(l)
l.pop(2)
print(l)
val = l.pop()
print(val)
print(l)
l.pop(0)
print(l)
#print(l.pop(1), l)
#print(l.pop(120), l)# IndexError: pop index out of range
#print([].pop()) # indexerro: pop from empty list
l.clear()
print(l)
del l
#print(l)#NameError: name 'l' is not defined
```

```python
# Attendance Management System
students = ["Lakshmi", "Ravi", "Sita"]
attendance = {}
# Mark attendance
for student in students:
    status = input(f"{student} (P/A): ")
    attendance[student] = status.upper()
# Display attendance
print("\nAttendance Report:")
for student in students:
    if attendance[student] == "P":
        print(student, "- Present")
    else:
        print(student, "- Absent")
```
