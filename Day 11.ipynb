```python
# Separate even and odd numbers from list of elements
# Input list
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
# numbers = [1,5,6,7,9,10,12]
even_numbers = []
odd_numbers = []
for num in numbers:
    if num % 2 == 0:
        even_numbers.append(num)
    else:
        odd_numbers.append(num)
print("Even numbers:", even_numbers)
print("Odd numbers:", odd_numbers)
```

```python
# find target element frequency
# [11,13,12,6,4,3,12,1,12]
# User-defined list
numbers = list(map(int, input("Enter elements separated by space: ").split()))
# User-defined target element
target = int(input("Enter target element: "))
# Count frequency
count = 0
for num in numbers:
    if num == target:
        count += 1
print("Frequency of", target, "is:", count)
```

```python
# remove the target element from array
# User-defined list
numbers = list(map(int, input("Enter elements separated by space: ").split()))
# User-defined target
target = int(input("Enter target element to remove: "))
# Remove all occurrences
new_list = []
for num in numbers:
    if num != target:
        new_list.append(num)
print("List after removing", target, ":", new_list)
```

```python
# find target element frequency
# by using functions
def find_frequency(arr, target):
    count = 0
    for num in arr:
        if num == target:
            count += 1
    return count


# User input
arr = list(map(int, input("Enter elements separated by space: ").split()))
target = int(input("Enter target element: "))

result = find_frequency(arr, target)

print("Frequency of", target, "is:", result)
```

```python
# remove the target element from array
# using functions
def remove_target(arr, target):
    new_arr = []
    for num in arr:
        if num != target:
            new_arr.append(num)
    return new_arr
# User input
arr = list(map(int, input("Enter elements separated by space: ").split()))
target = int(input("Enter target element to remove: "))
result = remove_target(arr, target)

print("Array after removing target:", result)
```
