```python
# Abstraction Method
# means hiding the internal details and showing only the essential features to the user
# Python provides abstraction using the abc module (Abstract Base Class)
from abc import ABC, abstractmethod
class Shape2D(ABC):
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass

class Shape3D(ABC):
    @abstractmethod
    def surface_area(self):
        pass
    @abstractmethod
    def volume(self):
        pass

class Cube(Shape3D):
    def __init__(self, side):
        self.side = side
    def surface_area(self):
        return 6 * self.side ** 2
    def volume(self):
        return self.side ** 3

print("3D Shapes")
cube = Cube(3)
print("Cube Surface_area:", cube.surface_area())
print("Cube volume:", cube.volume())
```

```python
# Shape
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def area(self):
        print("Area of Rectangle")

class Circle(Shape):
    def area(self):
        print("Area of Circle")

r = Rectangle()
r.area()

c = Circle()
c.area()
```

```python
# Vehicle
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Bike(Vehicle):
    def start(self):
        print("Bike starts with kick")

class Car(Vehicle):
    def start(self):
        print("Car starts with key")

b = Bike()
b.start()

c = Car()
c.start()
```

```python
# Bank System
# Different banks, same method → abstraction
from abc import ABC, abstractmethod

class Bank(ABC):

    @abstractmethod
    def interest_rate(self):
        pass

class SBI(Bank):
    def interest_rate(self):
        print("SBI Interest Rate: 6%")

class HDFC(Bank):
    def interest_rate(self):
        print("HDFC Interest Rate: 7%")

s = SBI()
s.interest_rate()

h = HDFC()
h.interest_rate()
```

```python
# Login System
from abc import ABC, abstractmethod

class Login(ABC):

    @abstractmethod
    def authenticate(self):
        pass

class EmailLogin(Login):
    def authenticate(self):
        print("Login using Email")

class GoogleLogin(Login):
    def authenticate(self):
        print("Login using Google")

e = EmailLogin()
e.authenticate()

g = GoogleLogin()
g.authenticate()
```

```python
# Payment Method
from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self):
        pass

class UPI(Payment):
    def pay(self):
        print("Pay using UPI")

class Card(Payment):
    def pay(self):
        print("Pay using Card")

p1 = UPI()
p1.pay()

p2 = Card()
p2.pay()
```

```python
# 2D & 3D Shapes (Abstraction)
from abc import ABC, abstractmethod
import math

# Abstract Class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# 2D Shapes 
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        print("Rectangle Area:", self.length * self.width)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print("Circle Area:", math.pi * self.radius * self.radius)

# 3D Shapes
class Sphere(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):   # Surface Area
        print("Sphere Surface Area:", 4 * math.pi * self.radius * self.radius)

class Cube(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):   # Surface Area
        print("Cube Surface Area:", 6 * self.side * self.side)
# Object Creation
r = Rectangle(5, 3)
r.area()

c = Circle(4)
c.area()

s = Sphere(3)
s.area()

cu = Cube(2)
cu.area()
```

```python
# 2D vs 3D with Area & Volume
from abc import ABC, abstractmethod
import math

# Abstract Class 
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# 2D Shapes
class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        print("Square Area:", self.side * self.side)

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        print("Triangle Area:", 0.5 * self.base * self.height)

# 3D Shapes
class Shape3D(Shape):

    @abstractmethod
    def volume(self):
        pass

class Cylinder(Shape3D):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def area(self):  # Surface Area
        print("Cylinder Surface Area:",
              2 * math.pi * self.radius * (self.radius + self.height))

    def volume(self):
        print("Cylinder Volume:",
              math.pi * self.radius * self.radius * self.height)

class Cuboid(Shape3D):
    def __init__(self, l, w, h):
        self.l = l
        self.w = w
        self.h = h

    def area(self):
        print("Cuboid Surface Area:",
              2 * (self.l*self.w + self.w*self.h + self.l*self.h))

    def volume(self):
        print("Cuboid Volume:", self.l * self.w * self.h)

# Objects 
s = Square(4)
s.area()

t = Triangle(6, 3)
t.area()

cy = Cylinder(3, 5)
cy.area()
cy.volume()

cb = Cuboid(2, 3, 4)
cb.area()
cb.volume()
```
