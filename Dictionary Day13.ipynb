```python
# dictionary values can acces by using keys
details = {'name':"venu","batch":49, "Course":"Python", 'name':'kalluri'}
print(details ['name'])
```

```python
# update dictionary value
details = {'name':"venu","batch":49, "Course":"Python", 'name':'kalluri',0:1}
details['Course']
print(details)
```

```python
# update dictionary value
details = {'name':"venu","batch":49, "Course":"Python", 'name':'kalluri',0:1}
details['Course'] = 'java'
print(details)
```

```python
# dictionary is mutable
# update dictionary values and items
details = {'name':"venu","batch":49, "Course":"Python", 'name':'kalluri',0:1}
details['Course'] = 'java'
details['age'] = 50
print(details)
```

```python
# dictionary methods or functions
# get(key, default_value) : it returns the values if the key is present in the dictionary
#              otherwise it returns default value
details = {'name':"venu","batch":49, "Course":"Python", 'name':'kalluri',0:1}
details['location'] #KeyError: 'location'
print(details.get('location','india'))
print(details.get('name', "Venu Royal"))
print(details.get('Marks'))
```

```python
# update(item):dictionary updates with new items
details = {'name':"venu","batch":49, "Course":"Python", 'name':'kalluri',0:1}
new_dict = {"course":"java", 'batch':50, 'location':'india'}
details.update(new_dict)
print(details)
```

```python
# keys(): it returns the list of dictionary keys
# values(): it returns the list of dictionary values
# items(): it returns the dictionary list of tuple of key-value pairs
details = {'name':"venu","batch":49, "Course":"Python", 'name':'kalluri',0:1}
print(details.keys())
print(details.values())
print(details.items())
print(details)
```

```python
# pop(key): it returns the values and removes the item from dictionary if key is present, otherwise KeyError
# popitem(): it return and removes the lastitem
details = {'name':"venu","batch":49, "Course":"Python", 'name':'kalluri',0:1}
print(details.pop('name'))
print(details.popitem())
print(details)
```

```python
phone_book = {}
def add_contact():
    name = input("Enter name: ")
    number = input("Enter phone number: ")
    phone_book[name] = number
    print("Contact added successfully")
def search_contact():
    name = input("Enter name to search: ")
    if name in phone_book:
        print("Phone Number:", phone_book[name])
    else:
        print("Contact not found")
def display_contacts():
    if len(phone_book) == 0:
        print("Phone book is empty")
    else:
        for name, number in phone_book.items():
            print(name, ":", number)
def delete_contact():
    name = input("Enter name to delete: ")
    if name in phone_book:
        del phone_book[name]
        print("Contact deleted")
    else:
        print("Contact not found")
while True:
    print("\n1.Add Contact")
    print("2.Search Contact")
    print("3.Display Contacts")
    print("4.Delete Contact")
    print("5.Exit")
    
    choice = int(input("Enter your choice: "))
    if choice == 1:
        add_contact()
    elif choice == 2:
        search_contact()
    elif choice == 3:
        display_contacts()
    elif choice == 4:
        delete_contact()
    elif choice == 5:
        print("Exiting...")
        break
    else:
        print("Invalid choice")
```

```python
phone_book = []
def add_contact():
    name = input("Enter name: ")
    number = input("Enter number: ")
    phone_book.append([name, number])
    print("Contact added")
def search_contact():
    name = input("Enter name to search: ")
    for contact in phone_book:
        if contact[0] == name:
            print("Phone Number:", contact[1])
            return
    print("Contact not found")
def display_contacts():
    for contact in phone_book:
        print(contact[0], ":", contact[1])
while True:
    print("\n1.Add  2.Search  3.Display  4.Exit")
    choice = int(input("Enter choice: "))
    if choice == 1:
        add_contact()
    elif choice == 2:
        search_contact()
    elif choice == 3:
        display_contacts()
    elif choice == 4:
        break
```

```python
# Phone Book program using functions
phone_book = {}
def add():
    name = input("Enter name: ")
    number = input("Enter number: ")
    phone_book[name] = number
    print("Contact added")
def search():
    name = input("Enter name to search: ")
    if name in phone_book:
        print("Number:", phone_book[name])
    else:
        print("Contact not found")
def display():
    for name, number in phone_book.items():
        print(name, ":", number)
while True:
    print("\n1.Add 2.Search 3.Display 4.Exit")
    ch = int(input("Enter choice: "))

    if ch == 1:
        add()
    elif ch == 2:
        search()
    elif ch == 3:
        display()
    elif ch == 4:
        break
```
